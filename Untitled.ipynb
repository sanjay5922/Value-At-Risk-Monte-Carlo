```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
initial_portfolio_value = 1_000_000  # Portfolio in dollars
mean_return = 0.0005                 # Daily expected return (0.05%)
volatility = 0.02                    # Daily volatility (2%)
days = 1                             # VaR over 1-day horizon
n_simulations = 100_000              # Number of simulations
```

```python
# Generate simulated daily returns
np.random.seed(42)
simulated_returns = np.random.normal(loc=mean_return, scale=volatility, size=n_simulations)
```

```python
# Calculate portfolio value changes
portfolio_changes = initial_portfolio_value * simulated_returns
```

```python
# Calculate Value-at-Risk at 95% and 99% confidence levels
VaR_95 = -np.percentile(portfolio_changes, 5)
VaR_99 = -np.percentile(portfolio_changes, 1)

print(f"95% 1-Day VaR: ${VaR_95:,.2f}")
print(f"99% 1-Day VaR: ${VaR_99:,.2f}")

# Plot histogram of portfolio P&L
plt.figure(figsize=(10,6))
plt.hist(portfolio_changes, bins=100, color='skyblue', edgecolor='black')
plt.axvline(-VaR_95, color='orange', linestyle='--', label='95% VaR')
plt.axvline(-VaR_99, color='red', linestyle='--', label='99% VaR')
plt.title('Simulated 1-Day Portfolio P&L Distribution')
plt.xlabel('Profit / Loss ($)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("var_histogram.png")
plt.show()
```
